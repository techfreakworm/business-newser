```python
import pandas as pd
import json

pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('source_list.csv')
```

```python
df.total.sum()
```

```python
df.head()
```

```python
df.columns = ['url', 'total']
```

```python
def get_url_str(val):
    try:
        return json.loads(val)['url']
    except:
        print('Error while getting url, base obj: ', val)
        return None
```

```python
df.url = df.url.apply(get_url_str)
```

```python
df.query('total > 14').sort_values(by='total',ascending=False).reset_index()
```
